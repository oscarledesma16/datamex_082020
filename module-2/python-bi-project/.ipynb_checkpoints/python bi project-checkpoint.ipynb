```python
import folium
import pandas as pd
```

```python
from pymongo import MongoClient
client=MongoClient()
db=client.companies
```

```python
colec = db.companies
colec
```

```python
# Decido filtrar la información para poder elegir una empresa de menos de 100 empleados y que haya sido fundada del 2008 
# en adelante y que su giro fuera sobre el análisis de datos.

list(colec.find({'$and':[{'number_of_employees':{'$lt':100}},
                   {'founded_year':{'$gt':2008}},
                {'category_code':{'$lt':'news'}}]},
                   {'_id':0, 'name':1, 'number_of_employees':1, 'description':1, 'founded_year':1}))
```

```python
# Selecciono la empresa de nombre Burt, que al investigar encuentro está ubicada en Suecia.

list(colec.find({'name':'Burt'},
                {'_id':0,'name':1, 'number_of_employees':1, 'description':1, 'founded_year':1}))
```

```python
# Tomo la decisión de que la rehubicación de la empresa sea en la capital, Estocolmo, en donde se concentra la mayor 
# actividad económica de este giro en este país.

mapa2=folium.Map(location=[59.342856, 18.051401], tiles='stamentoner', zoom_start=13)

mapa2
```

```python
# Se analisa la ciudad para determinar en qué punto sería el adecuado para colocar las nuevas oficinas, y se determina que la 
# mejor opción es en una ubicación que no sea tan céntrica pero que esté en una vialidad importante, así como de vías de 
# comunicación y transporte muy cerca.

m = folium.Map(
    location=[59.342856, 18.051401],
    zoom_start=15,
    tiles='Stamen Terrain'
)

marcador_empresa = folium.Marker(
    location=[59.342856, 18.051401], 
    tooltip='Lugar de empresa')

marcador_metro = folium.Marker(
    location=[59.342967, 18.049851],
    tooltip='Metro: Stockholm Odenplan',
    icon=folium.Icon(color="red"))

marcador_empresa.add_to(m)
marcador_metro.add_to(m)

m
```
